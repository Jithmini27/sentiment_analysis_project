```python
import pandas as pd

text = "I really love this product"
data = pd.DataFrame({"tweet": [text]})
```

```python
import re
import nltk
from nltk.corpus import stopwords

nltk.download("stopwords")
```

```python
data["tweet"] = data["tweet"].apply(
    lambda t: " ".join(word.lower() for word in str(t).split())
)

data["tweet"] = data["tweet"].apply(
    lambda t: re.sub(r'http\S+|www\S+', '', str(t))
)

data["tweet"] = data["tweet"].apply(
    lambda t: re.sub(r'[^\w\s]', '', str(t))
)

data["tweet"] = data["tweet"].apply(
    lambda t: re.sub(r'\d+', '', str(t))
)

stop_words = set(stopwords.words("english"))

data["tweet"] = data["tweet"].apply(
    lambda t: " ".join(word for word in str(t).split() if word not in stop_words)
)
```

```python
vocab = pd.read_csv("../static/model/vocabulary.txt", header=None)
tokens = vocab[0].tolist()
```

```python
def vectorizer(ds, vocabulary):
    vectorized_list = []

    for sentence in ds:
        sentence_lst = np.zeros(len(vocabulary))

        for i in range(len(vocabulary)):
            if vocabulary[i] in sentence.split():
                sentence_lst[i] = 1

        vectorized_list.append(sentence_lst)

    vectorized_list_new = np.asarray(vectorized_list, dtype=np.float32)
    return vectorized_list_new
```

```python
vectorized_text=vectorizer(data["tweet"],tokens)
```
